```python
a={'a':1,'b':3}
b={'b':1,'c':3}
d={}
for i,item in a.items():
    d[i]=item
for j,item in b.items():
  d[j]=d.get(j,0)+item
print(d)
```

```python
a={'a':1,'b':3,'c':1}
d={}
s=set()
for i,it in a.items():
  if it not in s:
    d[i]=it
    s.add(it)
print(d)
print(s)
```

```python
a={'a':1,'b':3}
b={'d':1,'c':3}
d={}
for i,item in a.items():
    d[i]=item
for i,item in b.items():
    d[i]=item
print(d)
```

```python
a={'a':1,'b':3,'c':1}
d={}
for i,v in a.items():
  d.setdefault(v,[]).append(i)
print(d)
```

```python
a={'a':1,'b':3}
b={'b':1,'c':3}
r={k for k in a if k in b}
print(r)
```

```python
a={'a':1,'b':3}
b={'b':1,'c':3}
r={k for k in a if k not in b}
print(r)
```

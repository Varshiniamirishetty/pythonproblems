```python
def gcd(a,b):
  if b==0:
    return a
  else:
    return gcd(b,a%b)
print(gcd(48,18))
```

```python
#functions
#recursions
n=12120
d=1
def count(n,d):
  if n==0:
    return 0
  if n%10==d:
    return 1+count(n//10,d)
  else:
    return count(n//10,d)
print(count(n,d))
```

```python
def reverse(n,r):
  if n==0:
    return r
  return reverse(n//10,r*10+n%10)
print(reverse(34,0))
```

```python
def sum(n,r):
  if n==0:
    return r
  return sum(n//10,r+n%10)
print(sum(121,0))
```

```python
def oneton(n):
  if n==0:
    return
  oneton(n-1)
  print(n)
oneton(5)
```

```python
def ntoone(n):
  if n==0:
    return
  print(n)
  ntoone(n-1)
ntoone(5)
```

```python
def palindrome(n,r=0):
  if n==0:
    return r
  return palindrome(n//10,r*10+n%10)
def is_palindrome(n):
  return n==palindrome(n)
print(is_palindrome(121))
```

```python
def sumofn(n):
  if n==0:
    return 0
  return n+sumofn(n-1)
print(sumofn(5))
```

```python
def maxele(n):
  if len(n)==1:
    return n[0]
  return max(n[0],maxele(n[1:]))
l=[2,9,3,1]
print(maxele(l))
```

```python
def lenofstr(s):
  if len(s)==0:
    return 0
  return 1+lenofstr(s[1:])
s='python'
print(lenofstr(s))
```

```python
n=12120
def count(n):
  if n==0:
    return 0
  return 1+count(n//10)
print(count(n))
```

```python
def dectobinary(n):
  if n==0:
    return 0
  return n%2+10*dectobinary(n//2)
print(dectobinary(10))
```
